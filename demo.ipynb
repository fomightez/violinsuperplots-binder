## Use the SuperViolin class to generate plots

You can also import the “superplot” class into your Python scripts. This section will give a sampling of that.

```python
# adapted from https://github.com/kynnemall/superviolin/blob/master/superviolin/superviolin/test_plot.py
from superviolin.plot import Superviolin
from superviolin import plot_cli
import pandas as pd
import matplotlib.pyplot as plt
# generate demo superplot
arguments = plot_cli.get_args(demonstration=True)
df = pd.read_csv("example_data/demo_data.csv")
superviolin_test = Superviolin(**arguments, dataframe=df)
print("Plotting superviolin")
superviolin_test.generate_plot()
ax = plt.gca()
```

This starts to allong more control than the command line interface offers. For example. We can show the plot again in the notebook as long as it is still define. Like so:

```python
ax.figure
```

We can save that plot object in various formats.

```python
ax.figure.savefig("demo.png")
```

```python
ax.figure.savefig("demo.svg")
```
