## Use the SuperViolin class to generate plots

You can also import the superplot class into your Python scripts. This section will give a sampling of that.

```python
# adapted from https://github.com/kynnemall/superviolin/blob/master/superviolin/superviolin/test_plot.py
from superviolin.plot import Superviolin
from superviolin import plot_cli
import pandas as pd
import matplotlib.pyplot as plt
# generate demo superplot
arguments = plot_cli.get_args(demonstration=True)
df = pd.read_csv("example_data/demo_data.csv")
superviolin_test = Superviolin(**arguments, dataframe=df)
print("Plotting superviolin")
superviolin_test.generate_plot()
p = plt.gca()
```

This starts to allong more control than the command line interface offers. For example. We can show the plot again in the notebook as long as it is still define. Like so:

```python
p.figure
```

### Saving the result

We can save that plot object in various image file formats.

```python
p.figure.savefig("demo.svg")
```

#### Portable network graphics format

Often it is easier to paste elsewhere using a form that can be easily viewed directly in many places, such as lossless `.png` format. The following will save it as `.png`.

```python
p.figure.savefig("demo.png")
```

Click on the Jupyter logo in the upper left to go to the Dashboard, and then click on the file in `.png` form listed to open. You'll see it larger now in its full glory.

**From that dashboard you can also then download image files of any plots produced to your local computer.**

You may need to adjust the size. With the vetor graphics SVG, you can do that in other software and it will scale well because vector graphics can be scaled without loss of resolution. With the `.png` format it is best to get closer to what you want as you make the image because it will become pixelated if you try to increase the image size in imaging software. Increasing the size of the image so that larger images still result in good resolution can be accomplished a couple ways while the plot is still in memory but after the code to generate it was run.  
One way is by saving the image at a higher resolution as demonstrated in the next cell.

```python
p.figure.savefig("larger_by_dpi_setting.png", dpi = (300))
```

Remember to go to the Dashboard, and then click on the file in `.png` form listed to open. You'll see it larger now in its full glory. The preview here won't let you appreciate the change.  

Another way us to set the size and then save the figure.

```python
p.figure.set_size_inches((7, 7))
p.figure.savefig("control_with_by_set_size.png")
```

Remember to go to the Dashboard, and then click on the file in `.png` form listed to open. You'll see it larger now in its full glory. The preview here won't let you appreciate the change.
